```python
num1 = input("Enter a number:")
num1 = float(num1)
num2 = input("Enter a second number:")
num2 = float(num2)

funct1=input("Enter a function:")

if funct1 == "+":
  print(num1 + num2)
elif funct1 =="-":
  print(num1-num2)
elif funct1== "*":
  print(num1*num2)
elif funct1== "/":
  print(num1/num2)
elif funct1== "**":
  print(num1**num2)
elif funct1== "%":
  print(num1%num2)
elif funct1== "//":
  print(num1//num2)
else:
  print("Invalid Operation")
```
